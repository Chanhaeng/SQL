### I assumed that utc_correction in users table tells you the right local time.
In that cae, I am going to divide table by morning (6-11a.m.), afternoon (11-5p.m.), night (6-12p.m.), and dawn (12-6a.m. and investigate the number of tests that are completed and divide it by date and month.

## 1. First assumption 
### divided by morning, afternoon, night, and dawn

a. When utc_correction applied

```python
%%sql
SELECT count(*) AS numtests,
       (CASE 
        WHEN(TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"11:00:00" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"05:59:59" THEN "Morning"
        
        WHEN(TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"17:00:00" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"10:59:59" THEN "Afternoon"
       
        WHEN((TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"23:59:59" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"16:59:59")
        OR 
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))="00:00:00"
        OR
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))="23:59:59"
        THEN "Evening"
        ELSE "DAWN"
        END) AS timelabel
FROM complete_tests c INNER JOIN (SELECT u.utc_correction AS utc_correction, d.dog_guid AS dog_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.utc_correction<0) a
    ON c.dog_guid=a.dog_guid
GROUP BY timelabel
ORDER BY numtests DESC;
```

b. when utc_correction not applied

```python
%%sql
SELECT TIME(c.created_at) AS time, count(*) AS numtests,
       (CASE 
        WHEN TIME(c.created_at)<"11:00:00" AND
        TIME(c.created_at)>"05:59:59" THEN "Morning"
        
        WHEN TIME(c.created_at)<"17:00:00" AND
        TIME(c.created_at)>"10:59:59" THEN "Afternoon"
       
        WHEN (TIME(c.created_at)<"23:59:59" AND
        TIME(c.created_at)>"16:59:59")
        OR 
        TIME(c.created_at)="00:00:00"
        OR
        TIME(c.created_at)="23:59:59"
        THEN "Evening"
        
        WHEN TIME(c.created_at)<"06:00:00" AND
        TIME(c.created_at)>"00:00:00" THEN "DAWN"
        END) AS timelabel
    
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.utc_correction<0) a
    ON c.dog_guid=a.dog_guid
GROUP BY timelabel
ORDER BY numtests DESC;
```

### There are a big difference between a and b (whether utc_correction is applied or not). 

When it is not applied, over 86% of the total complete tests are performed(tested) during night or dawn time, which rarely happen. 

When it is applied, there is a tendency found that people do the tests during afternoon or night time compared to morning time.

Then we are going to take a look by date.

```python
%%sql
SELECT DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)) AS day, count(*) AS numtests,
       (CASE
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=1 THEN "Su"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=2 THEN "Mo"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=3 THEN "Tu"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=4 THEN "We"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=5 THEN "Th"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=6 THEN "Fr"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=7 THEN "Sa"
        END) AS daylabel
FROM complete_tests c INNER JOIN (SELECT u.utc_correction AS utc_correction, d.dog_guid AS dog_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.utc_correction<0) a
    ON c.dog_guid=a.dog_guid
GROUP BY daylabel
ORDER BY numtests DESC;
```

```python
%%sql
SELECT DAYOFWEEK(c.created_at), COUNT(c.created_at) AS numtests,
(CASE
WHEN DAYOFWEEK(c.created_at)=1 THEN "Su"
WHEN DAYOFWEEK(c.created_at)=2 THEN "Mo"
WHEN DAYOFWEEK(c.created_at)=3 THEN "Tu"
WHEN DAYOFWEEK(c.created_at)=4 THEN "We"
WHEN DAYOFWEEK(c.created_at)=5 THEN "Th"
WHEN DAYOFWEEK(c.created_at)=6 THEN "Fr"
WHEN DAYOFWEEK(c.created_at)=7 THEN "Sa"
END) AS daylabel
FROM complete_tests c JOIN dogs d
ON c.dog_guid=d.dog_guid
WHERE d.exclude IS NULL OR d.exclude=0
GROUP BY daylabel
ORDER BY numtests DESC;
```

No matther whether utc_correction is applied or not, the results are the same. 

We are going to investigate it by date and time.

```python
%%sql
SELECT  count(*) AS numtests,
       (CASE 
        WHEN(TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"11:00:00" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"05:59:59" THEN "Morning"
        
        WHEN(TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"17:00:00" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"10:59:59" THEN "Afternoon"
       
        WHEN((TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))<"23:59:59" AND
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))>"16:59:59")
        OR 
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))="00:00:00"
        OR
        (TIME(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)))="23:59:59"
        THEN "Evening"
        ELSE "DAWN"
        END) AS timelabel,
        DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR)) AS day,
        (CASE
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=1 THEN "Su"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=2 THEN "Mo"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=3 THEN "Tu"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=4 THEN "We"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=5 THEN "Th"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=6 THEN "Fr"
        WHEN DAYOFWEEK(DATE_SUB(c.created_at, INTERVAL ABS(a.utc_correction) HOUR))=7 THEN "Sa"
        END) AS daylabel

FROM complete_tests c INNER JOIN (SELECT u.utc_correction AS utc_correction, d.dog_guid AS dog_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.utc_correction<0) a
    ON c.dog_guid=a.dog_guid
GROUP BY daylabel, timelabel
ORDER BY numtests DESC;
```

### As the table above suggests, we know that people do more tests on a certain date and during a certian period of time

```python
%%sql
SELECT a.country AS country, count(*) AS totaltests, COUNT(DISTINCT a.user_guid) AS numusers,
        count(*)/count(DISTINCT a.user_guid) AS avg
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid,
                                  u.state AS state, u.country As country
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                ) a
    ON  c.dog_guid=a.dog_guid
GROUP BY a.country
ORDER BY totaltests DESC;
```

### The fourth assumption

#### 4. Fee start user tend to complete less tests.

First, we are going to calculate the number of complete tests by a users. 

Like we saw before, the number of tests without rating is less than 10% of the total. (Complete_tests table is used for accuracy).

```python
%%sql
SELECT DISTINCT a.user_guid AS users, COUNT(*) AS numtest
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 ) a
    ON c.dog_guid=a.dog_guid
GROUP BY users
ORDER BY numtest DESC
;
```

Then how is the average by free start users?

```python
%%sql
SELECT distinct free_start_user as free, count(*) AS count
FROM users
GROUP BY free;
```

Free_start_user is indicated by None, 0 and 1. 
Let's assums that 1 is the one who started free.

```python
%%sql
SELECT DISTINCT a.user_guid AS users, COUNT(*) AS numtest
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.free_start_user=1) a
    ON c.dog_guid=a.dog_guid
GROUP BY users
ORDER BY numtest DESC
;
```

```python
%%sql
SELECT AVG(a.numtest) AS AVG
FROM (SELECT DISTINCT a.user_guid AS users, COUNT(*) AS numtest
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND u.free_start_user=1) a
    ON c.dog_guid=a.dog_guid
GROUP BY users) a
;
```

The average of users who is not free_start_users.

```python
%%sql
SELECT DISTINCT a.user_guid AS users, COUNT(*) AS numtest
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND ((u.free_start_user IS NULL) OR (u.free_start_user=0))) a
    ON c.dog_guid=a.dog_guid
GROUP BY users
ORDER BY numtest DESC
;
```

User whose id is "ce7b75bc-7144-11e5-ba71-058fbc01cf0b" completed too many tests compared to the others (considered as an outlier).

```python
%%sql
SELECT AVG(a.numtest) AS AVG
FROM (SELECT DISTINCT a.user_guid AS users, COUNT(*) AS numtest
FROM complete_tests c INNER JOIN (SELECT d.dog_guid AS dog_guid, u.user_guid AS user_guid
                                FROM dogs d INNER JOIN users u
                                    ON d.user_guid=u.user_guid
                                WHERE (d.exclude=0 OR d.exclude IS NULL) AND (u.exclude=0 OR u.exclude IS NULL)
                                 AND ((u.free_start_user IS NULL) OR (u.free_start_user=0)) ) a
    ON c.dog_guid=a.dog_guid
GROUP BY users
HAVING a.user_guid !="ce7b75bc-7144-11e5-ba71-058fbc01cf0b") a
;
```
